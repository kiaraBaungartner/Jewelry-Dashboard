```python
import pandas as pd
import numpy as np
```

```python
# Read Sales CSV
df_sales = pd.read_csv("sales_10000.csv")
```

```python
# Create random ratings: 1.0 to 5.0
df_sales["sales_rating"] = (
    np.clip(
        np.random.normal(loc=4.3, scale=0.5, size=len(df_sales)),
        1, 5
    )
    .round(1)
)
```

```python
# Read Products CSV 
df_products = pd.read_csv("products.csv")
```

```python
# Read Customer CSV 
df_customer = pd.read_csv("customers.csv")
```

```python
# Sales + Products + Customer Merge
df_merged = df_sales.merge(df_products, on="product_id", how="left") \
                    .merge(df_customer, on="customer_id", how="left")
```

```python
# Shows the first 5 rows of the merged dataframe
df_merged.head()
```

```python
# Export merged dataframe to CSV
df_merged.to_csv("sales.csv", index=False)
```
